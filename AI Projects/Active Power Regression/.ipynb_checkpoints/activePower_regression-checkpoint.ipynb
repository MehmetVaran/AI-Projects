```python
# importing libraries
import datetime
import time
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import plotly.express as px
import seaborn as sns
from spade import quit_spade
from spade.agent import Agent
from spade.behaviour import CyclicBehaviour, PeriodicBehaviour
from spade.message import Message
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from xgboost.sklearn import XGBRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import AdaBoostRegressor
```

```python
# reading datas and splitting into train test
data = pd.read_csv("T1.csv") # you can change the directory of csv file from here.
data = data[data['LV ActivePower (kW)'] > 0]
data = data.dropna()
date_time = data.iloc[:,0:1]
active_power = data.iloc[:,1:2]
wind_speed = data.iloc[:,2:3]
theoretical_power = data.iloc[:,3:4]
wind_direction = data.iloc[:,4:5]
x = pd.concat([wind_speed, theoretical_power], axis=1)
x = pd.concat([x,wind_direction],axis=1)
x_train, x_rem, y_train, y_rem = train_test_split(x,active_power, train_size=0.8, random_state=12)
x_valid, x_test, y_valid, y_test = train_test_split(x_rem,y_rem, test_size=0.5, random_state=12)
x_test_ws = x_test.iloc[:,0:1].values.tolist()
x_test_tp = x_test.iloc[:,1:2].values.tolist()
x_test_wd = x_test.iloc[:,2:3].values.tolist()
xgbRegressor = XGBRegressor()
```

```python
# creating ml agent that creates ml models,reports and predicts active power with best model
class MLAgent(Agent):
    class InformBehav(CyclicBehaviour):
        async def run(self):
            #getting test data
            msg = await self.receive(timeout=10)  # wait for a message for 10 seconds
            if msg:
                total_msg = msg.body
                msg_parts = total_msg.split("+")
                ws = float(msg_parts[0])
                tp = float(msg_parts[1])
                wd = float(msg_parts[2])
                list_msg={"Wind Speed (m/s)": [ws], "Theoretical_Power_Curve (KWh)": [tp], "Wind Direction (Â°)":[wd]}
                df_msg = pd.DataFrame(data=list_msg)
                self.data_predicted = self.data_predicted.append({"LV ActivePower (kW)":xgbRegressor.predict(df_msg)},ignore_index=True)
                print(self.data_predicted)
            else:
                mae_test = mean_absolute_error(y_test.iloc[0:26,:], self.data_predicted)
                print('Test Data MAE: %.3f' % mae_test)
                self.kill()
                
        async def on_end(self):
            # stop agent from behaviour
            await self.agent.stop()

        async def on_start(self):
            self.counter = 0
            self.data_predicted = pd.DataFrame(columns = ["LV ActivePower (kW)"])

    async def setup(self):
        print(f"MLAgent started at {datetime.datetime.now().time()}")
        start_at = datetime.datetime.now() + datetime.timedelta(seconds=5)
        # Training Models
        # Decision Tree
        Regressor = DecisionTreeRegressor()
        Regressor.fit(x_train, y_train)
        yhat_dt = Regressor.predict(x_valid)
        mae = mean_absolute_error(y_valid, yhat_dt)
        print('Decision Tree Regressor MAE: %.3f' % mae)
        R2_Decision_Tree = r2_score(y_valid,yhat_dt)
        print('Decision Tree Regressor R2: %.3f' %R2_Decision_Tree)
        ACC_Decision=Regressor.score(x_valid, y_valid)
        # Random Forest
        forest_model = RandomForestRegressor()
        forest_model.fit(x_train, y_train )
        yhat_rf = forest_model.predict(x_valid)
        mae = mean_absolute_error(y_valid, yhat_rf)
        print('Random Forest Regressor MAE: %.3f' % mae)
        R2_Random_Forest = r2_score(y_valid,yhat_rf)
        print('Random Forest Regressor R2: %.3f' %R2_Random_Forest)
        ACC_Random_Forest=Regressor.score(x_valid, y_valid)
        # XGBoost
        xgbRegressor.fit(x_train, y_train)
        yhat_xg = xgbRegressor.predict(x_valid)
        mae = mean_absolute_error(y_valid, yhat_xg)
        print('XGBoost Regressor MAE: %.3f' % mae)
        R2_XGB = r2_score(y_valid,yhat_xg)
        print('XGBoost Regressor R2: %.3f' %R2_XGB)
        ACC_XGB=xgbRegressor.score(x_valid, y_valid)
        # Extra Trees
        reg = ExtraTreesRegressor().fit(x_train, y_train)
        yhat_et = reg.predict(x_valid)
        mae = mean_absolute_error(y_valid, yhat_et)
        print('Extra Trees MAE: %.3f' % mae)
        R2_ExtraTrees = r2_score(y_valid,yhat_et)
        print('Extra Trees R2: %.3f' %R2_ExtraTrees)
        ACC_ExtraTrees=reg.score(x_valid, y_valid)
        # Ada Boost
        ada = AdaBoostRegressor()
        ada.fit(x_train , y_train)
        yhat_ada = ada.predict(x_valid)
        mae = mean_absolute_error(y_valid, yhat_ada)
        print('Ada Boost MAE: %.3f' % mae)
        R2_AdaBoost = r2_score(y_valid,yhat_ada)
        print('Ada Boost R2: %.3f' %R2_AdaBoost)
        ACC_Ada=ada.score(x_valid, y_valid)

        #plotting
        Error = [R2_Decision_Tree , R2_Random_Forest , R2_XGB , R2_ExtraTrees , R2_AdaBoost]
        dataPerf = pd.DataFrame(data={'Models': ['Decision_Tree', 'Random_Forest', 'XGB','ExtraTrees','AdaBoost'] ,
                                      'Error': Error ,
                                      'Accuracy':[ACC_Decision,ACC_Random_Forest,ACC_XGB,ACC_ExtraTrees,ACC_Ada]})

        plt.figure(figsize=(10, 8))
        sns.barplot(x="Models", y="Error", data=dataPerf, palette="magma")
        plt.title('Num of Models with their R2')
        plt.show()

        plt.figure(figsize=(10, 8))
        sns.barplot(x="Models", y="Accuracy", data=dataPerf, palette="magma")
        plt.title('Num of Models with their Accuracy')
        plt.show()
            
        b = self.InformBehav()
        self.add_behaviour(b)
```

```python
# creating manager agent that sends test values to ml agent
class ManagerAgent(Agent):
    class RecvBehav(PeriodicBehaviour):
        async def run(self):
            msg = Message(to="subs1@anonym.im")
            msg.body = str(x_test_ws[self.counter][0]) + "+" + str(x_test_tp[self.counter][0]) + "+" + str(x_test_wd[self.counter][0])
            await self.send(msg)
            print(f"{self.counter+1} data line is sent")
            
            if self.counter == 25:
                self.kill()
            self.counter += 1
            
        async def on_start(self):
            self.counter = 0
        async def on_end(self):
            await self.agent.stop()

    async def setup(self):
        print(f"ManagerAgent started at {datetime.datetime.now().time()}")
        start_at = datetime.datetime.now() + datetime.timedelta(seconds=5)
        b = self.RecvBehav(period=1, start_at=start_at)
        self.add_behaviour(b)
```

```python
if __name__ == "__main__":
    manageragent = ManagerAgent("subs2@anonym.im", "123456")
    mlagent = MLAgent("subs1@anonym.im","123456")

    
    future = mlagent.start(auto_register=True)
    future.result()
    
    manageragent.start(auto_register=True)


    while manageragent.is_alive():
        try:
            time.sleep(1)
        except KeyboardInterrupt:
            mlagent.stop()
            manageragent.stop()
            break
        print("Agents finished")
```
